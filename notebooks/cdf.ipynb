## Creating CDF

Given a random continuous function, I want to be able to calculate the CDF that will make it into a perfect uniform distribution

```python
from py_files.random_function import create_random_function
from py_files.pdf import make_pdf, visualize_pdf
from py_files.cdf import cdf_and_q, plot_cdf_transform
```

```python
#we are using some helper functions to create a random pdf; however, it is important to note that this pdf follows all of the requirements for a CDF to work for mapping to a perfect uniform distribution (it is continuous, positive, and integrates to 1)
p = make_pdf(create_random_function())
visualize_pdf(p)
```

```python
#p contains two things: the pdf itself and the samples used to create the pdf so we are extracting those two pieces 
pdf = p[1]
xs = p[0]
```

To find the CDF of a continuous random variable, you integrate its probability density function (PDF) from negative infinity to the variable x, resulting in the formula F(x) = P(X ≤ x) = ∫ from -∞ to x of f(t) dt. For different intervals of the PDF, you perform this integration for each interval, ensuring that the CDF is continuous and increases from 0 to 1. 

```python
u_samples, x_samples, u_check = cdf_and_q(pdf, xs)
```

```python
plot_cdf_transform(u_samples, x_samples, u_check, xs, pdf)
```
